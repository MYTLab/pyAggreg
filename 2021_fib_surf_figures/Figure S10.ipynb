```python
ch12_avg =  np.loadtxt('plot_data/Twist_angle/chain_12_00_relax_avg.txt')
ch24_avg =  np.loadtxt('plot_data/Twist_angle/chain_24_00_relax_avg.txt')
ch36_avg =  np.loadtxt('plot_data/Twist_angle/chain_36_00_relax_avg.txt')
ch48_avg =  np.loadtxt('plot_data/Twist_angle/chain_48_01_relax_avg.txt')
ch62_avg =  np.loadtxt('plot_data/Twist_angle/chain_62_00_relax_avg.txt')
```

```python
plt.figure(figsize=(20,8))
plt.title(r'Averaged twist angle ($\overline{\theta}$) ' ,fontsize=45)

plt.plot( ch12_avg, '-v', linewidth = 5, label = r'12 Chains' , markersize= 18, alpha= 0.6 )
plt.plot( ch24_avg, '-o', linewidth = 5, label = r'24 Chains' , markersize= 18, alpha= 0.6 )
plt.plot( ch36_avg, '-s', linewidth = 5, label = r'36 Chains' , markersize= 18, alpha= 0.6 )
plt.plot( ch48_avg, '-P', linewidth = 5, label = r'48 Chains' , markersize= 18, alpha= 0.6 )
plt.plot( ch62_avg, '-h', linewidth = 5, label = r'62 Chains' , markersize= 18, alpha= 0.6 )

plt.xlabel(r"Simulation time steps ( x $10^{4}$)", fontsize = 45)
plt.xticks(np.arange(101)[::10], fontsize = 35, rotation = 0)
plt.yticks(fontsize = 35)
plt.ylabel(r"$\overline{\theta}$", fontsize = 45, rotation=0)


# ##設定 label 的樣式
leg  = plt.legend()
leg_lines = leg.get_lines()
leg_texts = leg.get_texts()
plt.setp(leg_lines, linewidth=5)
plt.setp(leg_texts, fontsize=30)
plt.ylim(-3,2)
```

# Figure 6. (c)

```python
ch12_dis_01 =  np.loadtxt('plot_data/Twist_angle/chain_12_00_initial_dis.txt')
ch24_dis_01 =  np.loadtxt('plot_data/Twist_angle/chain_24_00_initial_dis.txt')
ch36_dis_01 =  np.loadtxt('plot_data/Twist_angle/chain_36_00_initial_dis.txt')
ch48_dis_01 =  np.loadtxt('plot_data/Twist_angle/chain_48_01_initial_dis.txt')
ch62_dis_01 =  np.loadtxt('plot_data/Twist_angle/chain_62_00_initial_dis.txt')
```

```python
plt.figure(figsize=(20,10))
## plot twist-angle distribution


sns.distplot(ch12_dis_01 , kde_kws={"label":"12 Chains" , "alpha": 0.5,"linewidth": 10} , hist= True)
sns.distplot(ch24_dis_01 , kde_kws={"label":"24 Chains" , "alpha": 0.5,"linewidth": 10} , hist= True)
sns.distplot(ch36_dis_01 , kde_kws={"label":"36 Chains" , "alpha": 0.5,"linewidth": 10} , hist= True)
sns.distplot(ch48_dis_01 , kde_kws={"label":"48 Chains" , "alpha": 0.5,"linewidth": 10} , hist= True)
sns.distplot(ch62_dis_01 , kde_kws={"label":"62 Chains" , "alpha": 0.5, "linewidth": 10} , hist= True)


plt.title(' Twist-angle distribution',fontsize=45)
plt.xlabel(r"$\theta$", fontsize = 50)
plt.xlim([-9,6])
plt.xticks(np.arange(-9,7,1), fontsize = 35)
plt.yticks(fontsize = 35)
plt.ylabel(r"$P$", fontsize = 50)

leg  = plt.legend()
leg_lines = leg.get_lines()
leg_texts = leg.get_texts()
plt.setp(leg_lines, linewidth=10)
plt.setp(leg_texts, fontsize=27)
```
