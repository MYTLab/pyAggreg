# Figure 4. (a)

```python
pdb_file1 = 'native_fibril_even_15A.pdb'
trj_file1 = 'movie_fibril_even_15A.dcd'
traj1 = md.load(trj_file1, top=pdb_file1, stride=1)
view1 = nv.show_mdtraj(traj1)
view1
```

```python
contact_traj1 = md.load(trj_file1, top=pdb_file1, stride=1)
topology1 = contact_traj1.topology
fibril1_CA_all = topology1.select("chainid 0 to 11 and name CA")
monomer1_CA = topology1.select("chainid 12 and name CA")
```

```python
fibril1_CA_chA = topology1.select("chainid 0  and name CA")     
fibril1_CA_chB = topology1.select("chainid 1  and name CA")      
fibril1_CA_chC = topology1.select("chainid 2  and name CA")      
fibril1_CA_chD = topology1.select("chainid 3  and name CA")      
fibril1_CA_chE = topology1.select("chainid 4  and name CA")      
fibril1_CA_chF = topology1.select("chainid 5  and name CA")      
fibril1_CA_chG = topology1.select("chainid 6  and name CA")      
fibril1_CA_chH = topology1.select("chainid 7  and name CA")      
fibril1_CA_chI = topology1.select("chainid 8  and name CA")      
fibril1_CA_chJ = topology1.select("chainid 9  and name CA")      
fibril1_CA_chK = topology1.select("chainid 10  and name CA")      
fibril1_CA_chL = topology1.select("chainid 11  and name CA")    
```

```python
traj1_contacts = ContactFrequency(contact_traj1[4102:],cutoff=0.65, haystack=fibril1_CA_all, query=monomer1_CA)
traj1_contacts1 = ContactFrequency(contact_traj1[4102:],cutoff=0.65, haystack=fibril1_CA_chA, query=monomer1_CA)
traj1_contacts2 = ContactFrequency(contact_traj1[4102:],cutoff=0.65, haystack=fibril1_CA_chB, query=monomer1_CA)
traj1_contacts3 = ContactFrequency(contact_traj1[4102:],cutoff=0.65, haystack=fibril1_CA_chC, query=monomer1_CA)
traj1_contacts4 = ContactFrequency(contact_traj1[4102:],cutoff=0.65, haystack=fibril1_CA_chD, query=monomer1_CA)
traj1_contacts5 = ContactFrequency(contact_traj1[4102:],cutoff=0.65, haystack=fibril1_CA_chE, query=monomer1_CA)
traj1_contacts6 = ContactFrequency(contact_traj1[4102:],cutoff=0.65, haystack=fibril1_CA_chF, query=monomer1_CA)
traj1_contacts7 = ContactFrequency(contact_traj1[4102:],cutoff=0.65, haystack=fibril1_CA_chG, query=monomer1_CA)
traj1_contacts8 = ContactFrequency(contact_traj1[4102:],cutoff=0.65, haystack=fibril1_CA_chH, query=monomer1_CA)
traj1_contacts9 = ContactFrequency(contact_traj1[4102:],cutoff=0.65, haystack=fibril1_CA_chI, query=monomer1_CA)
traj1_contacts10 = ContactFrequency(contact_traj1[4102:],cutoff=0.65, haystack=fibril1_CA_chJ, query=monomer1_CA)
traj1_contacts11 = ContactFrequency(contact_traj1[4102:],cutoff=0.65, haystack=fibril1_CA_chK, query=monomer1_CA)
traj1_contacts12 = ContactFrequency(contact_traj1[4102:],cutoff=0.65, haystack=fibril1_CA_chL, query=monomer1_CA)
```

```python
fig, ax = traj1_contacts.residue_contacts.plot(cmap='seismic', vmin=-1, vmax=1, figsize=(8, 6))
ax.set_xlim(*traj1_contacts.haystack_residue_range)
ax.set_ylim(*traj1_contacts.query_residue_range)
for i in np.arange(12):
    ax.vlines(0+i*32, 384, 416, colors = "blue", linestyles = "dashed")
ax.hlines(416, 0, 384, colors = "blue", linestyles = "dashed")
ax.hlines(416-6, 0, 384, colors = "blue", linestyles = "dashed")          #beta strand 3 have 7 residue
ax.hlines(416-10, 0, 384, colors = "blue", linestyles = "dashed")           #4 on loop2
ax.hlines(416-19, 0, 384, colors = "blue", linestyles = "dashed")           #9 on strand2
ax.hlines(416-24, 0, 384, colors = "blue", linestyles = "dashed")           #5 on loop1 
ax.hlines(416-31, 0, 384, colors = "blue", linestyles = "dashed")           #7 on strand1
plt.xlabel("Residue index of fibril", fontsize=18)
plt.ylabel("Residue index of incoming monomer", fontsize=18)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
```

```python
fig, ax = traj1_contacts1.residue_contacts.plot(cmap='seismic', vmin=-1, vmax=1, figsize=(8, 6))
ax.set_xlim(*traj1_contacts1.haystack_residue_range)
ax.set_ylim(*traj1_contacts1.query_residue_range)

for i in np.arange(416):
    ax.vlines(i, 0, 416, colors = "gray", linestyles = "dashed")
    ax.hlines(i, 0, 416, colors = "gray", linestyles = "dashed")

ax.hlines(416, 0, 384, colors = "blue", linestyles = "dashed")
ax.hlines(416-6, 0, 384, colors = "blue", linestyles = "dashed")          #beta strand 3 have 7 residue
ax.hlines(416-10, 0, 384, colors = "blue", linestyles = "dashed")           #4 on loop2
ax.hlines(416-19, 0, 384, colors = "blue", linestyles = "dashed")           #9 on strand2
ax.hlines(416-24, 0, 384, colors = "blue", linestyles = "dashed")           #5 on loop1 
ax.hlines(416-31, 0, 384, colors = "blue", linestyles = "dashed")           #7 on strand1
#plt.xlabel("Residue index of chain A", fontsize=18)
#plt.ylabel("Residue index of incoming monomer", fontsize=18)
plt.xticks(fontsize=0)
plt.yticks(fontsize=0)
```
