# Figure S9.

```python
ch12_tot_01 =  np.loadtxt('plot_data/Twist_angle/chain_12_00_01_tot.txt')
ch24_tot_01 =  np.loadtxt('plot_data/Twist_angle/chain_24_00_01_tot.txt')
ch36_tot_01 =  np.loadtxt('plot_data/Twist_angle/chain_36_00_01_tot.txt')
ch48_tot_01 =  np.loadtxt('plot_data/Twist_angle/chain_48_01_01_tot.txt')
ch62_tot_01 =  np.loadtxt('plot_data/Twist_angle/chain_62_00_01_tot.txt')
ch12_tot_00 =  np.loadtxt('plot_data/Twist_angle/chain_12_00_00_tot.txt')
ch24_tot_00 =  np.loadtxt('plot_data/Twist_angle/chain_24_00_00_tot.txt')
ch36_tot_00 =  np.loadtxt('plot_data/Twist_angle/chain_36_00_00_tot.txt')
ch48_tot_00 =  np.loadtxt('plot_data/Twist_angle/chain_48_01_00_tot.txt')
ch62_tot_00 =  np.loadtxt('plot_data/Twist_angle/chain_62_00_00_tot.txt')
ch12_tot_02 =  np.loadtxt('plot_data/Twist_angle/chain_12_00_02_tot.txt')
ch24_tot_02 =  np.loadtxt('plot_data/Twist_angle/chain_24_00_02_tot.txt')
ch36_tot_02 =  np.loadtxt('plot_data/Twist_angle/chain_36_00_02_tot.txt')
ch48_tot_02 =  np.loadtxt('plot_data/Twist_angle/chain_48_01_02_tot.txt')
ch62_tot_02 =  np.loadtxt('plot_data/Twist_angle/chain_62_00_02_tot.txt')
```

```python
plt.figure(figsize=(20,8))
plt.title(r'Accumulated total twist angle ($\theta_{tot}$) ' ,fontsize=45)

ch_12 = (np.array(ch12_tot_00)+np.array(ch12_tot_01)+np.array(ch12_tot_02))/3
ch_24 = (np.array(ch24_tot_00)+np.array(ch24_tot_01)+np.array(ch24_tot_02))/3
ch_36 = (np.array(ch36_tot_00)+np.array(ch36_tot_01)+np.array(ch36_tot_02))/3
ch_48 = (np.array(ch48_tot_00)+np.array(ch48_tot_01)+np.array(ch48_tot_02))/3
ch_62 = (np.array(ch62_tot_00)+np.array(ch62_tot_01)+np.array(ch62_tot_02))/3


####mean
ch12_err_tot = []
ch24_err_tot = []
ch36_err_tot = []
ch48_err_tot = []
ch62_err_tot = []

for i in np.arange(len(ch_12)):
    ch12_err = np.std([np.array(ch12_tot_00)[i], np.array(ch12_tot_01)[i], np.array(ch12_tot_02)[i]], ddof=1)
    ch24_err = np.std([np.array(ch24_tot_00)[i], np.array(ch24_tot_01)[i], np.array(ch24_tot_02)[i]], ddof=1)
    ch36_err = np.std([np.array(ch36_tot_00)[i], np.array(ch36_tot_01)[i], np.array(ch36_tot_02)[i]], ddof=1)
    ch48_err = np.std([np.array(ch48_tot_00)[i], np.array(ch48_tot_01)[i], np.array(ch48_tot_02)[i]], ddof=1)
    ch62_err = np.std([np.array(ch62_tot_00)[i], np.array(ch62_tot_01)[i], np.array(ch62_tot_02)[i]], ddof=1)
    ch12_err_tot.append(ch12_err)
    ch24_err_tot.append(ch24_err)
    ch36_err_tot.append(ch36_err)
    ch48_err_tot.append(ch48_err)
    ch62_err_tot.append(ch62_err)
plt.errorbar(np.arange(101)[::5], ch_12[::5], ch12_err_tot[::5], fmt='o', color='blue', ecolor='blue', capsize=8, elinewidth=3)
plt.errorbar(np.arange(101)[::5], ch_24[::5], ch24_err_tot[::5], fmt='o', color='orange', ecolor='orange', capsize=8, elinewidth=3)
plt.errorbar(np.arange(101)[::5], ch_36[::5], ch36_err_tot[::5], fmt='o', color='green', ecolor='green', capsize=8, elinewidth=3)
plt.errorbar(np.arange(101)[::5], ch_48[::5], ch48_err_tot[::5], fmt='o', color='red', ecolor='red', capsize=8, elinewidth=3)
plt.errorbar(np.arange(101)[::5], ch_62[::5], ch62_err_tot[::5], fmt='o', color='purple', ecolor='purple', capsize=8, elinewidth=3)

plt.plot( ch_12, linewidth = 5, label = r'12 Chains' , markersize= 18, alpha= 0.6 )
plt.plot( ch_24, linewidth = 5, label = r'24 Chains' , markersize= 18, alpha= 0.6 )
plt.plot( ch_36, linewidth = 5, label = r'36 Chains' , markersize= 18, alpha= 0.6 )
plt.plot( ch_48, linewidth = 5, label = r'48 Chains' , markersize= 18, alpha= 0.6 )
plt.plot( ch_62, linewidth = 5, label = r'62 Chains' , markersize= 18, alpha= 0.6 )


plt.xlabel(r"Simulation time steps ( x $10^{6}$)", fontsize = 45)
plt.xticks(np.arange(101)[::10], fontsize = 35, rotation = 0)
plt.yticks(fontsize = 35)
plt.ylabel(r"$\theta_{tot}$", fontsize = 45)


# ##設定 label 的樣式
leg  = plt.legend()
leg_lines = leg.get_lines()
leg_texts = leg.get_texts()
plt.setp(leg_lines, linewidth=5)
plt.setp(leg_texts, fontsize=22)
plt.ylim(-300,30)
```
