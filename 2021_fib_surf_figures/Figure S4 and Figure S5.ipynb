```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import seaborn as sns
```

# Figure S4. PMF (fibril-like ensemble)

```python
plot_data = np.loadtxt('plot_data/PMF/PMF_k10_fibril_back.txt')
```

```python
distance = []
pmf = []
for i in np.arange(len(plot_data)):
    distance.append(plot_data[i][0])
    pmf.append(plot_data[i][1])
```

```python
x_pre = np.array(distance)
y_pre = np.array(pmf)
y = y_pre - y_pre.max()
fig, ax = plt.subplots(figsize=(9, 6))
ax.plot(x_pre, y,'s')
ax.vlines(4.9, 0, -20, colors = "red", linestyles = "dashed")
ax.vlines(7.6, 0, -20, colors = "red", linestyles = "dashed")
ax.set_xlabel(r"$r$ (nm)", fontsize=24)
ax.set_ylabel(r"Free energy (kcal/mol)", fontsize=24)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.xlim(0,10.5)
ax.set_title("Potential of mean force", fontsize=25)
```

# Figure S5. Distribution of sampling windows (fibril-like ensemble)

```python
dis_plot_data = np.loadtxt('plot_data/PMF/fibril_back_sampling_distribution.txt')
```

```python
plt.subplots(figsize=(20,10))
for i in range(10):
    ax = sns.distplot(dis_plot_data[i], norm_hist=True, kde_kws=dict(linewidth=4), color='black')
    ax = sns.distplot(dis_plot_data[10+i], norm_hist=True, kde_kws=dict(linewidth=4), color='blue')
    ax = sns.distplot(dis_plot_data[20+i], norm_hist=True, kde_kws=dict(linewidth=4), color='red')
    ax = sns.distplot(dis_plot_data[30+i], norm_hist=True, kde_kws=dict(linewidth=4), color='orange')
    ax = sns.distplot(dis_plot_data[40+i], norm_hist=True, kde_kws=dict(linewidth=4), color='purple')
    ax = sns.distplot(dis_plot_data[50+i], norm_hist=True, kde_kws=dict(linewidth=4), color='gray')
    ax = sns.distplot(dis_plot_data[60+i], norm_hist=True, kde_kws=dict(linewidth=4), color='g')
    ax = sns.distplot(dis_plot_data[70+i], norm_hist=True, kde_kws=dict(linewidth=4), color='y')
    ax = sns.distplot(dis_plot_data[80+i], norm_hist=True, kde_kws=dict(linewidth=4), color='aqua')
ax = sns.distplot(dis_plot_data[90], norm_hist=True, kde_kws=dict(linewidth=4), color='aqua')
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
ax.set_xlabel("distance(nm)", fontsize=25)
ax.set_ylabel("density", fontsize=25)
ax.set_title("900 frames sampling distribution", fontsize=30)
ax.legend(('window 1-10','window 11-20','window 21-30','window 31-40','window 41-50','window 51-60','window 61-70','window 71-80','window 81-91') ,fancybox=True, fontsize=22)
```
